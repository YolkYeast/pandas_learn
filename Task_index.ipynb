## 一、单级索引

### 1. loc、iloc、[]操作符

- iloc:位置索引
- loc:标签索引
- []？

#### (a)loc方法

**①单行索引**

```python
df.loc[1102]
```

**②多行索引**

```python
df.loc[[1103, 1105]]
```

注意：所有在loc中使用的切片全部包含**右端点**！这是因为如果作为Pandas的使用者，那么肯定不太关心最后一个标签再往后一位是什么，但是如果是左闭右开，那么就很麻烦，先要知道再后面一列的名字是什么，非常不方便，因此Pandas中将loc设计为**左右全闭**）

```python
df.loc[1304: 2103].head()
```

```python
df.loc[2402::-1].head()
```

**③单列索引**

```python
df.loc[:, 'Height'].head()
```

```python
df.loc[:,'Height':'Math'].head()
```

**⑤联合索引:**

```python
df.loc[1102:2401:3, "Height":"Math"].head()
```

**⑥函数式索引**

```python
df.loc[lambda x:x["Gender"]=="M"].head()
```

```python
#这里的例子表示，loc中能够传入函数，并且函数的输入值是整张表，输出为标量、切片、合法列表（元素出现在索引中）、合法索引
def f(x):
    return [1101,1103]
df.loc[f]
```

**⑦布尔索引:**

```python
df.loc[df['Address'].isin(['street_7', 'street_4'])].head()
```

```python
df.loc[[True if i[-1]=='4' or i[-1]=='7' else False for i in df['Address'].values]].head()
```

**小节：本质上说，loc中能传入的只有布尔列表和索引子集构成的列表，只要把握这个原则就很容易理解上面那些操作**

#### (b)iloc方法(与loc不同，**切点右端点不包括**)

**①单行索引**

```python
df.iloc[3]
```

**②多行索引:**

**切点右端不包括**

```python
df.iloc[3:6]
```

**③单列索引:**

```python
df.iloc[:, 3].head()
```

**④多列索引:**

**右端不包括哟**

```python
df.iloc[:, 3:6:2].head()
```

**⑤混合索引:**

```python
df.iloc[3::4, 7::-2].head()
```

**⑥函数式索引:**

```python
df.iloc[lambda x : [3]].head()
```

**小节：iloc中接收的参数只能为整数或整数列表或布尔列表，不能使用布尔Series，如果要用就必须如下把values拿出来**

```python
#df.iloc[df['School']=='S_1'].head() #报错
```

```python
df.iloc[(df['School'] == 'S_1').values].head()
```
